```python
import os
import pandas as pd

# Project root (adjust if needed)
project_root = r"C:\Users\kalki\OneDrive\Desktop\week2\mobile-banking-reviews-Challenge-week2"

# Paths
processed_path = os.path.join(project_root, "notebooks", "data", "processed")
data_path = os.path.join(project_root, "notebooks", "data")

# Load Task 1 raw/cleaned reviews
task1 = pd.read_csv(os.path.join(processed_path, "reviews_processed.csv"))

# Load Task 2 sentiment + theme output
task2 = pd.read_csv(os.path.join(data_path, "task2_sentiment_themes.csv"))  # <- correct folder

print("Task 1 columns:", task1.columns)
print("Task 2 columns:", task2.columns)

# Merge by review_id (common field)
df = task1.merge(task2, on="review_id", how="left")

print("Merged columns:", df.columns)

# Save final dataset
final_csv_path = os.path.join(processed_path, "reviews_final.csv")
df.to_csv(final_csv_path, index=False)

print(f"FINAL CSV CREATED → {final_csv_path}")
```

```python
import pandas as pd
import os
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon
nltk.download('vader_lexicon')

# Paths
processed_path = r"C:\Users\kalki\OneDrive\Desktop\week2\mobile-banking-reviews-Challenge-week2\notebooks\data\processed"
final_csv_path = os.path.join(processed_path, "reviews_final.csv")

# Create folder if it doesn't exist
os.makedirs(processed_path, exist_ok=True)

# Load merged CSV (Task 1 + Task 2)
df = pd.read_csv(final_csv_path)  # or your current merged CSV path

# Ensure we have a 'review_text' column
if 'review_text' not in df.columns:
    for alt in ('review_text_x', 'review_text_y'):
        if alt in df.columns:
            df['review_text'] = df[alt]
            break
    else:
        raise KeyError(f"No review text column found. Available columns: {list(df.columns)}")

# Initialize VADER
sia = SentimentIntensityAnalyzer()

# Compute sentiment scores and labels
df['sentiment_score'] = df['review_text'].fillna('').apply(lambda x: sia.polarity_scores(str(x))['compound'])
df['sentiment_label'] = df['sentiment_score'].apply(
    lambda score: 'positive' if score > 0.05 else ('negative' if score < -0.05 else 'neutral')
)

# Save final dataset
df.to_csv(final_csv_path, index=False)
print(f"Sentiment added → {final_csv_path}")
```
